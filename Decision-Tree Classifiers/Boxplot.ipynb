```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DM_A-2_secondpart.csv')
df= data.iloc[:,1:6]
gini=list(df.iloc[:,0])
chi= list(df.iloc[:,1])
info=list(df.iloc[:,2])
gain= list(df.iloc[:,3])
misclassify=list(df.iloc[:,4])

plot_dat=[gini,chi,info,gain,misclassify]

plt.style.use('seaborn')
#plt.grid(color='white', linestyle='solid')

fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
#ax.set(facecolor = "orange")


bp = ax.boxplot(plot_dat, patch_artist = True,meanline=True)

colors = ['#FFCCFF', '#FFC0CB', 
          '#FFD700', '#DDA0DD',"#33CCFC"]  
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

for whisker in bp['whiskers']:
    whisker.set(color ='red',
                linewidth = 1.5,
                linestyle =":")
    
ax.set_xticklabels(['gini', 'chisquare', 
                    'infogain', 'gain_ratio','misclassify'],fontsize=16)

plt.title("Model box plot")
  
plt.savefig('boxplot.png')
```
